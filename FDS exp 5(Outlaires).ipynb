```python
import numpy as np
array=np.random.randint(1,100,16) 
def outDetection(array):
    sorted(array)
```

```python
def outDetection(array):
    sorted(array)
    Q1,Q3=np.percentile(array,[25,75])
    IQR=Q3-Q1
    lr=Q1-(1.5*IQR)
    ur=Q3+(1.5*IQR)
    return lr,ur
lr,ur=outDetection(array)
```
